```python
from math import radians, cos, sin, asin, sqrt

import copy
```

```python
def distance(x, y): #naive distance 
    x1,x2,y1,y2 = float(x[0]),float(x[1]),float(y[0]),float(y[1])
    return sqrt((y2 - x2)**2 + (y1 - x1)**2)

def haversine_distance(x, y): #may take long
    #implement haversine formula to account for Earth's curvature
      
    lat1,lon1,lat2,lon2 = float(x[0]),float(x[1]),float(y[0]),float(y[1])
    
    # convert decimal degrees to radians 
    lon1, lat1, lon2, lat2 = map(radians, [lon1, lat1, lon2, lat2])

    # haversine formula 
    dlon = lon2 - lon1 
    dlat = lat2 - lat1 
    a = sin(dlat/2)**2 + cos(lat1) * cos(lat2) * sin(dlon/2)**2
    c = 2 * asin(sqrt(a)) 
    m = 3956 # Radius of earth in miles. Use 3956 for miles
    return c * m

def oneSwap(S,kmeans=0,f=0,k=0):
        origCost = cost(S,kmeans,f,k)
        t = .99
        for ind, y in enumerate(C):
            for ind2, x in enumerate(S):
                if x == ind:
                    continue
                newS = copy.deepcopy(S)
                newS[ind2] = ind
                if(cost(newS,kmeans,f,k)<t*origCost):
                    return True, newS
        return False,S

def cost(S,kmeans=0,f=0,k=0):
    cost = 0
    for ind, y in enumerate(C): #find the closest s in S for each c in C
        minD = float('inf')
        for x in S: #distance = 0 
            if x == ind:
                continue
            if G[ind][x] < minD:
                minD = G[ind][x]
        cost+= minD
    return cost if kmeans == 0 else cost + k*f
```
